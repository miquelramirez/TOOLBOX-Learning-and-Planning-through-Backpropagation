```python
%matplotlib inline
import random
import os
import json
import sys
import string
import unicodedata
import pandas as pd
import seaborn as sns
import itertools
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Given local path, find full path
def PathFinder(path):
    script_dir = os.path.dirname('__file__')
    fullpath = os.path.join(script_dir,path)
    return fullpath

#Read Data for Deep Learning
def ReadData(path):
    fullpath=PathFinder(path)
    return np.loadtxt(fullpath,delimiter=',')

def getScales(X,Y):
    distances = np.sqrt(np.power(X-5,2)+np.power(Y-5,2))
    scalefactor = 2.0/(1.0+np.exp(-2*distances))-0.99;
    return scalefactor
```

```python
hvac_heuristic = ReadData('../HVAC_PLOT_SOURCE/HVAC_INTERM_Heuristic.csv')
hvac_tensorflow = ReadData('../HVAC_PLOT_SOURCE/HVAC_INTERM_Tensorflow.csv')
```

```python
def get_frame(data):
    row,column = data.shape
    new_data = []
    for i in range(row-1):
        level=0
        for j in range(column):
            #if j%12 == 0:
                #level+=1
            #room = j%12
            new_data.append([int(i),int(j+1),data[i,j]])
    return pd.DataFrame(np.array(new_data),columns=['Time_Step','Room','Temperature'])
```

```python
heuristic = get_frame(hvac_heuristic)
heuristic.Time_Step = heuristic.Time_Step.astype(int)
heuristic.Room = heuristic.Room.astype(int)
heuristic=heuristic[heuristic['Time_Step']<19]
tensorflow = get_frame(hvac_tensorflow)
tensorflow.Time_Step = tensorflow.Time_Step.astype(int)
tensorflow.Room = tensorflow.Room.astype(int)
tensorflow=tensorflow[tensorflow['Time_Step']<19]
```

```python
heuristic['Method'] = 'Heuristic'
tensorflow['Method'] = 'Tensorflow'
df=pd.concat([tensorflow,heuristic])
df.head()
```

```python
sns.set(style="white",  palette="colorblind", color_codes=True)
fig,ax = plt.subplots(figsize=(12, 3), dpi=100)
ax.set_autoscaley_on(True)
sns.violinplot(x="Time_Step", y="Temperature",size=1,hue="Method",palette={"Tensorflow": "b", "Heuristic": "r"},split=True,linewidth=1, inner=None, data=df)
plt.axhline(y=21.75,ls=":", c="k")
plt.axhline(y=23.5,ls=":", c="0.5")
plt.axhline(y=20.0,ls=":", c="0.5")
fig.get_axes()[0].legend(loc='lower right')
plt.savefig('../PLOTS/hvac_tensorflow_heuristic_violin.eps',format='eps')
```

```python
hvac_reward_heuristic = ReadData('../HVAC_PLOT_SOURCE/HVAC_REWARD_Heuristic.csv')
hvac_reward_tensorflow = ReadData('../HVAC_PLOT_SOURCE/HVAC_REWARD_Tensorflow.csv')
```

```python
def get_reward_frame(data):
    row = len(data)
    new_data = []
    for i in range(1,row):
        new_data.append([int(i), -data[i]])
    return pd.DataFrame(np.array(new_data),columns=['Time_Step','Loss'])
```

```python
reward_heuristic = get_reward_frame(hvac_reward_heuristic)
reward_tensorflow = get_reward_frame(hvac_reward_tensorflow)
```

```python
reward_heuristic['Method'] = 'Heuristic'
reward_tensorflow['Method'] = 'Tensorflow'
```

```python
reward_heuristic.Time_Step = reward_heuristic.Time_Step.astype(int)
reward_tensorflow.Time_Step = reward_tensorflow.Time_Step.astype(int)
```

```python
df_reward=pd.concat([reward_tensorflow,reward_heuristic])
df_reward.head()
```

```python
sns.set(style="white",  palette="colorblind", color_codes=True)
fig,ax = plt.subplots(figsize=(12, 3), dpi=100)
ax.set_yscale('log')
sns.barplot(x="Time_Step", y="Loss", hue="Method", palette={"Tensorflow": "b", "Heuristic": "r"},data=df_reward)
ax.set_ylabel("Cost")
fig.get_axes()[0].legend(loc='upper right')
plt.savefig('../PLOTS/hvac_tensorflow_heuristic_action.eps',format='eps')
```
