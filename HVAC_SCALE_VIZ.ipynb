```python
%matplotlib inline
import random
import os
import json
import sys
import string
import unicodedata
import pandas as pd
import seaborn as sns
import itertools
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Given local path, find full path
def PathFinder(path):
    script_dir = os.path.dirname('__file__')
    fullpath = os.path.join(script_dir,path)
    return fullpath

#Read Data for Deep Learning
def ReadData(path):
    fullpath=PathFinder(path)
    return np.loadtxt(fullpath,delimiter=',')

def getScales(X,Y):
    distances = np.sqrt(np.power(X-5,2)+np.power(Y-5,2))
    scalefactor = 2.0/(1.0+np.exp(-2*distances))-0.99;
    return scalefactor
```

```python
hvac_data = ReadData('HVAC_INTERM.csv')
```

```python
def get_frame(data):
    row,column = data.shape
    new_data = []
    for i in range(row-1):
        level=0
        for j in range(column):
            #if j%12 == 0:
                #level+=1
            #room = j%12
            new_data.append([int(i),int(j+1),data[i,j]])
    return pd.DataFrame(np.array(new_data),columns=['Time_Step','Room','Temperature'])
```

```python
df = get_frame(hvac_data)
df.head()
df.Time_Step = df.Time_Step.astype(int)
df.Room = df.Room.astype(int)
df=df[df['Time_Step']<9]
```

```python
fig = plt.figure(figsize=(8, 3), dpi=300)
sns.violinplot(x="Time_Step", y="Temperature",size=1, data=df)
```
